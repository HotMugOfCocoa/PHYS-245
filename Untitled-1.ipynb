```python
import seaborn as sns
import matplotlib.pyplot as plt

def create_2d_histogram(x_data, y_data, x_label, y_label, title):
    """Create a 2D histogram using Seaborn
    
    Parameters:
    x_data (array-like): x-axis data
    y_data (array-like): y-axis data
    x_label (str): label for x-axis
    y_label (str): label for y-axis
    title (str): plot title
    
    Returns:
    None
    """
    # create plot
    sns.histplot(x=x_data, y=y_data, cbar=True, cmap="YlGnBu")
    
    # set labels and title
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    
    # show plot
    plt.show()
```

```python
import numpy as np

# create some example data
x_data = np.random.normal(size=1000)
y_data = np.random.normal(size=1000)

# create the plot using the function
create_2d_histogram(x_data, y_data, "X Label", "Y Label", "2D Histogram Example")
```
